Ejercicio4 representa algunos datos en un gráfico de barras

La r en una cadena indica que se trata de una cadena raw. Las cadenas raw se distinguen de las normales en que los caracteres escapados mediante la barra invertida ( \ ) no se sustituyen por sus contrapartidas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

df= pd.read_json(r"https://raw.githubusercontent.com/AdrianPlaza17/Examen-2---Actividad-plot---json/main/sample2.json")                                                                                                  

print(df)
nombres =[]
pesos=[]
for i in df.index:
    nombres.append(df["nombre"][i])
    pesos.append(df["peso"][i])



plt.pie(pesos, labels=nombres)
plt.show()


 

plt.bar(nombres, pesos)
plt.xlabel('Nombre')
plt.ylabel('Peso (kg)')
plt.title('Peso de cada persona')

plt.show()
```
